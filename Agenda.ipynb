```python
class Contacto:
    def __init__(self, nombre, apellido, telefono, direccion, empleo):
        self.nombre = nombre
        self.apellido = apellido
        self.telefono = telefono
        self.direccion = direccion
        self.empleo = empleo

class Agenda:
    def __init__(self):
        self.contactos = []

    def crear_contacto(self):
        print("\nCreación de un nuevo contacto:")
        nombre = input("Nombre: ").strip()
        apellido = input("Apellido: ").strip()
        telefono = input("Teléfono: ").strip()
        direccion = input("Dirección: ").strip()
        empleo = input("Empleo: ").strip()

        if nombre and apellido and telefono and direccion and empleo:
            nuevo_contacto = Contacto(nombre, apellido, telefono, direccion, empleo)
            self.contactos.append(nuevo_contacto)
            print("Contacto creado exitosamente.")
        else:
            print("Error: Todos los campos son obligatorios. Intente nuevamente.")

    def buscar_y_modificar_contacto(self):
        termino_busqueda = input("\nIngrese el término de búsqueda: ").lower()

        resultados = [contacto for contacto in self.contactos if
                      (termino_busqueda in contacto.nombre.lower() or
                       termino_busqueda in contacto.apellido.lower() or
                       termino_busqueda in contacto.telefono or
                       termino_busqueda in contacto.direccion.lower() or
                       termino_busqueda in contacto.empleo.lower())]

        if resultados:
            print("\nResultados de la búsqueda:")
            for i, contacto in enumerate(resultados, 1):
                print(f"{i}. {contacto.nombre} {contacto.apellido} - Teléfono: {contacto.telefono}")

            seleccion = input("\nSeleccione el número del contacto que desea modificar (0 para cancelar): ")
            if seleccion.isdigit():
                indice_contacto = int(seleccion) - 1
                if 0 <= indice_contacto < len(resultados):
                    self.modificar_contacto(resultados[indice_contacto])
                else:
                    print("Selección inválida.")
            else:
                print("Entrada inválida.")
        else:
            print("No se encontraron contactos que coincidan con la búsqueda.")

    def modificar_contacto(self, contacto):
        print(f"\nModificación de contacto: {contacto.nombre} {contacto.apellido}")
        print("1. Modificar Nombre")
        print("2. Modificar Apellido")
        print("3. Modificar Teléfono")
        print("4. Modificar Dirección")
        print("5. Modificar Empleo")
        print("6. Eliminar Contacto")
        opcion = input("Seleccione el campo que desea modificar (1/2/3/4/5/6): ")

        if opcion == '1':
            nuevo_nombre = input("Nuevo Nombre: ")
            contacto.nombre = nuevo_nombre
        elif opcion == '2':
            nuevo_apellido = input("Nuevo Apellido: ")
            contacto.apellido = nuevo_apellido
        elif opcion == '3':
            nuevo_telefono = input("Nuevo Teléfono: ")
            contacto.telefono = nuevo_telefono
        elif opcion == '4':
            nueva_direccion = input("Nueva Dirección: ")
            contacto.direccion = nueva_direccion
        elif opcion == '5':
            nuevo_empleo = input("Nuevo Empleo: ")
            contacto.empleo = nuevo_empleo
        elif opcion == '6':
            confirmacion = input(f"¿Está seguro que desea eliminar a {contacto.nombre} {contacto.apellido}? (s/n): ")
            if confirmacion.lower() == 's':
                self.contactos.remove(contacto)
                print("Contacto eliminado exitosamente.")
            else:
                print("Operación de eliminación cancelada.")
        else:
            print("Opción inválida.")

    def mostrar_todos_los_contactos(self):
        if self.contactos:
            print("\nTodos los contactos:")
            for i, contacto in enumerate(self.contactos, 1):
                print(f"{i}. {contacto.nombre} {contacto.apellido} - Teléfono: {contacto.telefono}")
        else:
            print("No hay contactos registrados.")

    def mostrar_menu_principal(self):
        print("\n--- Agenda de Contactos ---")
        print("1. Crear Contacto")
        print("2. Buscar y Modificar Contacto")
        print("3. Mostrar Todos los Contactos")
        print("4. Salir")

    def ejecutar(self):
        while True:
            self.mostrar_menu_principal()
            opcion = input("Seleccione una opción: ")

            if opcion == '1':
                self.crear_contacto()
            elif opcion == '2':
                self.buscar_y_modificar_contacto()
            elif opcion == '3':
                self.mostrar_todos_los_contactos()
            elif opcion == '4':
                print("Gracias por usar la Agenda de Contactos.")
                break
            else:
                print("Opción inválida. Por favor, seleccione una opción válida.")

# Ejecutar la Agenda
agenda = Agenda()
agenda.ejecutar()
```
